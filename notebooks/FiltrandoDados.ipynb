```python
import pandas as pd
```

```python
df = pd.read_parquet(url_raw)
```

```python
fl_date = pd.to_datetime(df['FL_DATE'], format='%m/%d/%Y %I:%M:%S %p')
period_begin = fl_date.min().strftime('%Y-%m-%d')
period_end = fl_date.max().strftime('%Y-%m-%d')
print(f"\nPeriod of time:")
print(f"Begin date: {period_begin}")
print(f"End date: {period_end}")
```

```python
# Filter flights with either origin or destination city name containing "New York"
ny_flights = df[df['ORIGIN_CITY_NAME'].str.contains('New York') | df['DEST_CITY_NAME'].str.contains('New York')]

# Print the filtered flight information
print(ny_flights.shape)
```
